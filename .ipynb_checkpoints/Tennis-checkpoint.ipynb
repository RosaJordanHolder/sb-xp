```python
import pandas as pd
import numpy as np
import os
import itertools
import random
```

```python
homedir = os.path.expanduser("~")
folder = os.path.join(homedir, "Documents\\GitHub\\tennis_atp")
```

```python
def BigDF(folder):
    csv_list = []
    files = [os.path.join(folder, file) for file in os.listdir(folder) if len(file.split("_")) == 3 and "matches" in file]
    for file in files:
        csv = pd.read_csv(os.path.join(folder, file))
        csv_list.append(csv)
    DF = pd.concat(csv_list, sort = False)
    DF.index = range(len(DF))
    
    return DF
```

```python
DF = BigDF(folder)
DF = DF.dropna(subset = ["w_svpt", "w_1stIn", "l_svpt", "l_1stIn"], how = "any")
```

```python
w_cols1 = [n for n in DF.columns if n.split("_")[0] == "w"]
w_cols2 = [n for n in DF.columns if n.split("_")[0] == "winner"]
l_cols1 = [n for n in DF.columns if n.split("_")[0] == "l"]
l_cols2 = [n for n in DF.columns if n.split("_")[0] == "loser"]
```

```python
p_cols11 = [n.replace("w_", "p_") for n in w_cols1]
p_cols12 = [n.replace("winner", "player") for n in w_cols2]
o_cols21 = [n.replace("l_", "o_") for n in l_cols1]
o_cols22 = [n.replace("loser", "opponent") for n in l_cols2]

p_cols21 = [n.replace("l_", "p_") for n in l_cols1]
p_cols22 = [n.replace("loser", "player") for n in l_cols2]
o_cols11 = [n.replace("w_", "o_") for n in w_cols1]
o_cols12 = [n.replace("winner", "opponent") for n in w_cols2]
```

```python
d1 = dict(zip(w_cols1+w_cols2+l_cols1+l_cols2, p_cols11+p_cols12+o_cols21+o_cols22))
d2 = dict(zip(l_cols1+l_cols2+w_cols1+w_cols2, p_cols21+p_cols22+o_cols11+o_cols12))
```

```python
df = pd.concat([DF.rename(columns = d1), DF.rename(columns = d2)], sort = False)
```

```python
df["year"] = df["tourney_date"]//10000
df["month"] = (df["tourney_date"] - df["year"] * 10000)//100
df["day"] = df["tourney_date"] - df["year"] * 10000 - df["month"] * 100

df["total_points"] = df["p_svpt"] + df["o_svpt"]
```

```python
gb = df.groupby(["player_name"]).sum().reset_index()
```

```python
gb["1stSv%"] = gb["p_1stIn"] / gb["p_svpt"]
gb["1stSvWin%"] = gb["p_1stWon"] / gb["p_1stIn"]
gb["2ndSv%"] = (gb["p_svpt"] - gb["p_1stIn"] - gb["p_df"])/(gb["p_svpt"] - gb["p_1stIn"])
gb["2ndSvWin%"] = gb["p_2ndWon"] / (gb["p_svpt"] - gb["p_1stIn"] - gb["p_df"])

gb["rtn1stSvWin%"] = 1 - gb["o_1stWon"] / gb["o_1stIn"]
gb["rtn2ndSvWin%"] = 1 - gb["o_2ndWon"] / (gb["o_svpt"] - gb["o_1stIn"] - gb["o_df"])
```

```python
gb[gb.total_points>20000].sort_values("rtn1stSvWin%", ascending = False)
```

```python
def point(server, returner, dataframe):
    serv = dataframe.loc[dataframe.player_name == server]
    retu = dataframe.loc[dataframe.player_name == returner]
    
    server_pt = [1, 0]
    ret_pt = [0, 1]
    
    r = np.random.rand(4)
    if r[0] < serv["1stSv%"].values[0]:
        if r[1] < ((serv["1stSvWin%"].values[0] + 1 - retu["rtn1stSvWin%"].values[0])/2):
            return server_pt
        else:
            return ret_pt
    elif r[2] < serv["2ndSv%"].values[0]:
        if r[3] < (serv["2ndSvWin%"].values[0] + 1 - retu["rtn2ndSvWin%"].values[0])/2:
            return server_pt
        else:
            return ret_pt
    else:
        return ret_pt
```

```python
def Game(server, returner, dataframe):
    score = [0, 0]
    
    test = True
    while test:
        p = point(server, returner, dataframe)
        
        score[0] += p[0]
        score[1] += p[1]
        
        diff = score[0] - score[1]
        if max(score) == 4 and abs(diff) > 1:
            test = False
        elif score[0] >= 2 and score[1] >= 2 and abs(diff) > 1:
            test = False
    
    if diff > 0:
        return server
    else:
        return returner
```

```python
def Tiebreak(first, second, dataframe):
    score = [0, 0]
    
    i, j = 0, 0
    s, r = first, second
    test = True
    while test:
        if abs(i-j) > 0:
            s, r = r, s
        if s == first:
            i += 1
            x = 0
        else:
            j += 1
            x = 1
        p = point(s, r, dataframe)
        score[0] += p[x]
        score[1] += p[1-x]
        diff = score[0] - score[1]
        
        if max(score) == 7 and abs(diff) > 1:
            test = False
        elif score[0] >= 6 and score[1] >= 6 and abs(diff) > 1:
            test = False
    
    if diff > 0:
        return first
    else:
        return second
```

```python
def Set(first, second, dataframe, tiebreak = True):
    score = {first: 0, second: 0}
    
    s, r = first, second
    test = True
    while test:
        g = Game(server = s, returner = r, dataframe = dataframe)
        score[g] += 1
        s, r = r, s
        
        diff = score[first] - score[second]
        
        if max(score.values()) == 6 and abs(diff) > 1:
            test = False
        elif score[first] == score[second] and score[first] == 6 and tiebreak == True:
            t = Tiebreak(s, r, dataframe)
            score[t] += 1
            test = False
        elif max(score.values()) > 6 and abs(diff) > 1:
            test = False
            
    return score
```

```python
def Match(first, second, dataframe, sets=5, tiebreaks = True, final_tiebreak = True):
    score = []
    sets_won = []
    
    i = 0
    f, s = first, second
    while i < sets:
        if i == sets - 1:
            t = Set(f, s, dataframe, tiebreak = final_tiebreak)
        else:
            t = Set(f, s, dataframe, tiebreak = tiebreaks)
        t["Set"] = i + 1
        t["Set Winner"] = list(t.keys())[list(t.values()).index(max(t.values()))]
        score.append(t)

        i += 1
        if (t[first] + t[second])%2 == 1:
            f, s = s, f
        
        if t[first] > t[second]:
            sets_won.append(first)
        else:
            sets_won.append(second)
        
        for val in set(sets_won):
            if sets_won.count(val) >= sets/2:
                return score  
            
    return score
```

```python
def Tournament(entrants, dataframe, sets = 5, tiebreaks = True, final_tiebreak = True):
    named_rounds = {8: "Quarter Finals", 4: "Semi Finals", 2: "Final"}

    j = 0
    test = True
    through = entrants
    while test:
        print("##########################")
        j += 1
        if j == 1:
            random.shuffle(through)
        pairs = [(i, (through[2*i], through[2*i + 1])) for i in range(len(through)//2)]
        r = j
        if len(through) in named_rounds.keys():
            r = named_rounds[len(through)]
        through = []
        for pair in pairs:
            contestants = list(pair[1])
            random.shuffle(contestants)
            m = Match(first = contestants[0], second = contestants[1], 
                      dataframe=dataframe, sets=sets, tiebreaks = tiebreaks, 
                      final_tiebreak = final_tiebreak)
            mdf = pd.DataFrame(m)
            vc = mdf["Set Winner"].value_counts()
            winner = vc.loc[vc==max(vc)].index.values[0]
            through.append(winner)
            
            non_set_cols = [n for n in mdf.columns if "Set" not in n]
            print("Round {}:".format(r), non_set_cols)
            print(mdf[non_set_cols].values.tolist())
            print("Winner:", winner)
            print()
        
        if len(through) == 1:
            print("CHAMPION")
            test = False
    
    return through
```

```python
gb_filt = gb.loc[gb.total_points > 8000]
players = gb_filt.player_name.unique()
combs = itertools.combinations(players, 2)
entrants = random.sample(list(players), 128)

Tournament(entrants, dataframe = gb, final_tiebreak = False)
```
