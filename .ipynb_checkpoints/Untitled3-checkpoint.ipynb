```python
import os
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
import matplotlib.patches as pch
```

```python
homedir = os.path.expanduser("~")
dataFolder = os.path.join(homedir, "Documents",  "Github", "open-data", "data")
```

```python
competitions = os.path.join(dataFolder, "competitions.json")
events = os.path.join(dataFolder, "events")
lineups = os.path.join(dataFolder, "lineups")
matches = os.path.join(dataFolder, "matches")
```

```python
def timestamptoseconds(timestamp):
    broken = timestamp.split(":")
    s = sum([float(broken[i]) * (60 ** (2 - i)) for i in range(len(broken))])
    return s

def timestamptominutes(timestamp):
    m = timestamptoseconds(timestamp) / 60
    return m    

def tts(timestamps):
    ts = [timestamptoseconds(n) for n in timestamps]
    return ts

def ttm(timestamps):
    ts = [timestamptominutes(n) for n in timestamps]
    return ts
```

```python
def positivenegative(a):
    if a > 0:
        return 1
    elif a < 0:
        return -1
    else:
        return 0
```

```python
def json_loads(filepath):
    with open(filepath, encoding="utf-8") as json_file:
        return json.load(json_file)
```

```python
def json_dataFrame(filepath):
    json_file = json_loads(filepath)
    df = pd.json_normalize(json_file)
    
    check_columns = {"period", "timestamp"}
    if check_columns.issubset(df.columns): 
        x = df.groupby(["period"])["timestamp"].last()
        x[0] = "00:00:00"
        for i in x.index[:-1]:
            df.loc[df["period"] == i, "time"] = [n + timestamptominutes(x[i-1]) for n in ttm(df.loc[df["period"] == i, "timestamp"])] 
    return df
```

```python
competitionsDataFrame = json_dataFrame(competitions)
```

```python
allmatchesDataFrame = pd.DataFrame()

for comp_index, comp_row in competitionsDataFrame.iterrows():
    
    path = os.path.join(matches, str(comp_row["competition_id"]), str(comp_row["season_id"]) + ".json")
    matchesDataFrame = json_dataFrame(path)
    allmatchesDataFrame = pd.concat([allmatchesDataFrame, matchesDataFrame])
    
    for match_index, match_row in matchesDataFrame.iterrows():
        
        path2 = os.path.join(events, str(match_row["match_id"]) + ".json")
        eventsDataFrame = json_dataFrame(path2)
        print(comp_row["competition_name"], comp_row["season_name"], match_row["match_id"])
        break
        
#eventsDataFrame
allmatchesDataFrame.index = range(len(allmatchesDataFrame))
```

```python
def draw_pitch(pitch_color = "green", line_color = "white"):
    fig, ax = plt.subplots(figsize = (12, 8))
    ax.axis("off")
    fig.patch.set_facecolor(pitch_color)
    for side in ["top", "bottom", "left", "right"]:
        ax.spines[side].set_visible(False)   
    
    x1, x2 = -5, 125
    y1, y2 = -5, 85
    z = 1
    
    plt.xlim(x1, x2)
    plt.ylim(y1, y2)
    
    Pitch = pch.Rectangle([0, 0], width = 120, height = 80, fill = False, color = line_color, zorder = z)
    
    Halfway = pch.ConnectionPatch([60, 0], [60, 80], "data", "data", color = line_color, zorder = z)
    CentreCircle = pch.Circle((60, 40), 10, color = line_color, fill = False, zorder = z)
    CentreSpot = pch.Circle((60, 40), 0.6, color = line_color, fill = True, zorder = z)
    
    LeftPenalty = pch.Rectangle([0, 18], width = 18, height = 44, color = line_color, fill = False, zorder = z)
    RightPenalty = pch.Rectangle([102, 18], width = 18, height = 44, color = line_color, fill = False, zorder = z)
    
    LeftPenSpot = pch.Circle((12, 40), 0.6, color = line_color, fill = True, zorder = z)
    RightPenSpot = pch.Circle((108, 40), 0.6, color = line_color, fill = True, zorder = z)
    
    LeftSixYard = pch.Rectangle([0, 30], width = 6, height = 20, color = line_color, fill = False, zorder = z)
    RightSixYard = pch.Rectangle([114, 30], width = 6, height = 20, color = line_color, fill = False, zorder = z)
    
    LeftGoal = pch.Rectangle([-3, 36], width = 3, height = 8, color = line_color, fill = False, zorder = z)
    RightGoal = pch.Rectangle([120, 36], width = 3, height = 8, color = line_color, fill = False, zorder = z)
    
    LeftArc = pch.Arc((12, 40), height = 17.8, width = 17.8, angle = 0, theta1 = 310, theta2 = 50, color = line_color, zorder = z)
    RightArc = pch.Arc((108, 40), height = 17.8, width = 17.8, angle = 0, theta1 = 130, theta2 = 230, color = line_color, zorder = z)
    
    for i in [Pitch, Halfway, CentreCircle, CentreSpot,
              LeftPenalty, RightPenalty, LeftPenSpot, RightPenSpot,
              LeftSixYard, RightSixYard, LeftGoal, RightGoal, LeftArc, RightArc]:
        ax.add_patch(i)
        
    return fig, ax, plt
    
```

```python
allmatchesDataFrame.index = range(len(allmatchesDataFrame))
matches_number = len(allmatchesDataFrame)
random_match = np.random.randint(0, matches_number)
path2 = os.path.join(events, str(allmatchesDataFrame.loc[random_match, "match_id"]) + ".json")
eventsDataFrame = json_dataFrame(path2)
path2
```

```python
fig, ax, plt = draw_pitch(pitch_color = "white", line_color = "black")

teams = eventsDataFrame["team.name"].unique()
X = [0, 120]
Y = [80, 0]
markers = ["o", "s", "x"]
cmap = "rainbow"

totals = {}

for i in [0, 1]:
    team = teams[i]
    oppo = teams[int((i + 1)%2)]
    color = "C" + str(i)
    
    tests = {"shots": (eventsDataFrame["team.name"] == team) & (eventsDataFrame["type.name"] == "Shot"),
        "goals": (eventsDataFrame["team.name"] == team) & (eventsDataFrame["type.name"] == "Shot") & (eventsDataFrame["shot.outcome.name"] == "Goal"), 
        "ogs": (eventsDataFrame["team.name"] == oppo) & (eventsDataFrame["type.name"] == "Own Goal Against")}
    
    minis = {}
    for key in tests.keys():
        for period in [[1, 2, 3, 4], [5]]:
            if period[0] == 1:
                new_key = key
            else:
                new_key = "_".join([key, "shootout"])
            minis[new_key] = eventsDataFrame.loc[tests[key] & (eventsDataFrame["period"].isin(period)), ["shot.statsbomb_xg", "location", "player.name", "time"]] 

    test_names = ["shots", "goals", "ogs"]
    for j in range(len(test_names)):
        mini = minis[test_names[j]]
        x = [abs(n[0] - X[i]) for n in mini["location"]]
        y = [abs(n[1] - Y[i]) for n in mini["location"]]
        if test_names[j] == "ogs":
            c = "black"
        else:
            c = mini["shot.statsbomb_xg"] / 0.8
        plt.scatter(x, y, s = 90, c = c, marker = markers[j], zorder = 2, cmap = cmap, alpha = 0.4, edgecolor = "black")
    
    if len(minis["shots_shootout"]) == 0:
        gs = len(minis["goals"]) + len(minis["ogs"])
    else:
        gs = str((len(minis["goals"]) + len(minis["ogs"]))) + " (" + str(len(minis["goals_shootout"])) + ")"
        
    placeDict = {"v": [91, team], "Score": [87, gs], "Shots": [84, len(minis["shots"])], "xG Total": [81, round(minis["shots"]["shot.statsbomb_xg"].sum(), 2)]}
    for key in placeDict.keys():
        if key == "v": 
            f = 12
            c = "black"
        else: 
            f = 10
            c = "black"
        plt.text(60 - (i - 0.5) * 30, placeDict[key][0], str(placeDict[key][1]), horizontalalignment = "center", fontsize = f, color = c)
    
    totals[team] = minis
    
for key in placeDict.keys():
    plt.text(60, placeDict[key][0], key, horizontalalignment = "center")
    

plt.show()
        
```
