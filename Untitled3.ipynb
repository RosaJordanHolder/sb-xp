```python
def tts(timestamp):
    broken = timestamp.split(":")
    s = sum([float(broken[i]) * (60 ** (2 - i)) for i in range(len(broken))])
    return s

def ttm(timestamp):
    m = tts(timestamp) / 60
    return m    
```

```python
def positivenegative(a):
    if a > 0:
        return 1
    elif a < 0:
        return -1
    else:
        return 0
```

```python
def json_loads(filepath):
    with open(filepath, encoding="utf-8") as json_file:
        return json.load(json_file)
```

```python
def locationFinder(row):
    end_cols = [n for n in row.index if "end_location" in n]
    for col in end_cols:
        if isinstance(row[col], list):
            return row[col]
    return row[col]
```

```python
def locationFinder1(row):
    xs = row.values
    for x in xs:
        if isinstance(x, list):
            return x
```

```python
def json_dataFrame1(filepath):
    json_file = json_loads(filepath)
    df = pd.json_normalize(json_file)
    
    check_columns = {"location", "period", "timestamp"}
    if check_columns.issubset(df.columns): 
        x = df.groupby(["period"])["timestamp"].last()
        x[0] = "00:00:00"
        vfunc = np.vectorize(ttm)
        y = vfunc(x.sort_index().values).cumsum()
        for i in x.index[:(-1)]:
            df.loc[df["period"] == i, "time"] = [ttm(n) + y[i-1] for n in df.loc[df["period"] == i, "timestamp"].values.tolist()]
        
        df["end_location"] = df[[n for n in df.columns if "end_location" in n]].dropna(thresh = 1).apply(locationFinder1, axis = 1)
        
        for pre in ["end_", ""]:
            if pre == "end_":
                suffs = ["x", "y", "z"]
            else:
                suffs = ["x", "y"]
            cols = ["{}location_{}".format(pre, suff) for suff in suffs]
            lox = df["{}location".format(pre)].dropna()
            
            df[cols] = pd.DataFrame(lox.tolist(), columns = cols, index = lox.index)
            
        fna = {"pass.outcome.name": "Complete", "pass.cross": False, "pass.type.name": "Regular"}
        df.loc[df["type.name"] == "Pass", :] = df.loc[df["type.name"] == "Pass", :].fillna(fna)
        
        #Pressures, axis = 1
        pdf = df.loc[df["type.name"] == "Pressure", ["time", "possession"]]
        for row in pdf.itertuples():
            t = row[1] + 1/12
            rdf = df.loc[(df["time"] >= row[1]) & (df["time"] <= t), "possession"].max()
            if rdf > row[2]:
                df.at[row[0], "pressure_regain"] = 1
        
        #Progress
        tests = {}
        tests["progress1"] = (df["location_x"] >= 60) & (df["end_location_x"] - df["location_x"] >= 10)
        tests["deepprogress"] = (df["location_x"] <= 80) & (df["end_location_x"] > 80)
        for prog in ["Pass", "Carry"]:
            if prog != "Carry":
                test1 = (df["type.name"] == prog) & (df[prog.lower() + ".outcome.name"] == "Complete")
            else:
                test1 = df["type.name"] == prog
                
            for key in tests.keys():
                test2 = test1 & tests[key]

                df.loc[test1, "_".join([prog, key])] = 1
         
        match_id = os.path.basename(filepath).split(".json")[0]
        df.loc[:, "match_id"] = match_id
        
    return df
```

```python
def json_dataFrame(filepath):
    json_file = json_loads(filepath)
    df = pd.json_normalize(json_file)
    
    check_columns = {"location", "period", "timestamp"}
    if check_columns.issubset(df.columns): 
        x = df.groupby(["period"])["timestamp"].last()
        x[0] = "00:00:00"
        vfunc = np.vectorize(ttm)
        y = vfunc(x.sort_index().values).cumsum()
        for i in x.index[:(-1)]:
            df.loc[df["period"] == i, "time"] = [ttm(n) + y[i-1] for n in df.loc[df["period"] == i, "timestamp"].tolist()]
        
        elCols = [n for n in df.columns if "end_location" in n]
        els = [df[col].dropna() for col in elCols]
        df["end_location"] = pd.concat(els)
        
        for pre in ["end_", ""]:
            if pre == "end_":
                suffs = ["x", "y", "z"]
            else:
                suffs = ["x", "y"]
            cols = ["{}location_{}".format(pre, suff) for suff in suffs]
            lox = df["{}location".format(pre)].dropna()
            
            df[cols] = pd.DataFrame(lox.tolist(), columns = cols, index = lox.index)
            
        fna = {"pass.outcome.name": "Complete", "pass.cross": False, "pass.type.name": "Regular"}
        df.loc[df["type.name"] == "Pass", :] = df.loc[df["type.name"] == "Pass", :].fillna(fna)
        
        #Pressures, axis = 1
        pressures = df["type.name"] == "Pressure"
        poss2 = df.groupby("possession")["time"].first().tolist()
        timetest = df["time"] == -1
        for t in poss2:
            timetest |= (df["time"] >= t - 1/12) & (df["time"] <= t)
        
        df.loc[pressures & timetest, "pressure_regain"] = 1
        
        #Progress
        tests = {}
        tests["progress1"] = (df["location_x"] >= 60) & (df["end_location_x"] - df["location_x"] >= 10)
        tests["deepprogress"] = (df["location_x"] <= 80) & (df["end_location_x"] > 80)
        for prog in ["Pass", "Carry"]:
            if prog != "Carry":
                test1 = (df["type.name"] == prog) & (df[prog.lower() + ".outcome.name"] == "Complete")
            else:
                test1 = df["type.name"] == prog
                
            for key in tests.keys():
                test2 = test1 & tests[key]

                df.loc[test1, "_".join([prog, key])] = 1
         
        match_id = os.path.basename(filepath).split(".json")[0]
        df.loc[:, "match_id"] = match_id
        
    return df
```

```python
def NameColumn(row):
    if row["player_nickname"] == None:
        name = row["player_name"]
    else:
        name = row["player_nickname"]
    return name.strip()
```

```python
def ShortName(name):
    split = name.split(" ")
    l = len(split)
    if l == 1:
        return name
    elif l >= 2:
        if "van" in split:
            s = split.index("van")
        else:
            s = -1
        return split[0][0] + ". " + " ".join(split[s:])
```

```python
def draw_pitch(pitch_color = "white", line_color = "black", y_size = 10):
    x_size = y_size * 1.5
    fig, ax = plt.subplots(figsize = (x_size, y_size))
    ax.axis("off")
    fig.patch.set_facecolor(pitch_color)
    for side in ["top", "bottom", "left", "right"]:
        ax.spines[side].set_visible(False)   
    
    x1, x2 = -5, 125
    y1, y2 = -5, 85
    z = 1
    
    plt.xlim(x1, x2)
    plt.ylim(y1, y2)
    
    Pitch = pch.Rectangle([0, 0], width = 120, height = 80, fill = False, color = line_color, zorder = z)
    
    Halfway = pch.ConnectionPatch([60, 0], [60, 80], "data", "data", color = line_color, zorder = z)
    CentreCircle = pch.Circle((60, 40), 10, color = line_color, fill = False, zorder = z)
    CentreSpot = pch.Circle((60, 40), 0.3, color = line_color, fill = True, zorder = z)
    
    LeftPenalty = pch.Rectangle([0, 18], width = 18, height = 44, color = line_color, fill = False, zorder = z)
    RightPenalty = pch.Rectangle([102, 18], width = 18, height = 44, color = line_color, fill = False, zorder = z)
    
    LeftPenSpot = pch.Circle((12, 40), 0.3, color = line_color, fill = True, zorder = z)
    RightPenSpot = pch.Circle((108, 40), 0.3, color = line_color, fill = True, zorder = z)
    
    LeftSixYard = pch.Rectangle([0, 30], width = 6, height = 20, color = line_color, fill = False, zorder = z)
    RightSixYard = pch.Rectangle([114, 30], width = 6, height = 20, color = line_color, fill = False, zorder = z)
    
    LeftGoal = pch.Rectangle([-3, 36], width = 3, height = 8, color = line_color, fill = False, zorder = z)
    RightGoal = pch.Rectangle([120, 36], width = 3, height = 8, color = line_color, fill = False, zorder = z)
    
    LeftArc = pch.Arc((12, 40), height = 17.8, width = 17.8, angle = 0, theta1 = 310, theta2 = 50, color = line_color, zorder = z)
    RightArc = pch.Arc((108, 40), height = 17.8, width = 17.8, angle = 0, theta1 = 130, theta2 = 230, color = line_color, zorder = z)
    
    for i in [Pitch, Halfway, CentreCircle, CentreSpot,
              LeftPenalty, RightPenalty, LeftPenSpot, RightPenSpot,
              LeftSixYard, RightSixYard, LeftGoal, RightGoal, LeftArc, RightArc]:
        ax.add_patch(i)
        
    return fig, ax
```

```python
def LineupDataFrame(match_id):
    lpath = os.path.join(lineups, str(match_id) + ".json")
    lineup = json_loads(lpath)
    llist = []
    for i in range(len(lineup)):
        df = pd.json_normalize(lineup[i]["lineup"])
        df.loc[:, "team.name"] = lineup[i]["team_name"]
        df.loc[:, "team.id"] = lineup[i]["team_id"]
        llist.append(df)
    ldf = pd.concat(llist)

    ldf.loc[:, "Name"] = ldf.apply(NameColumn, axis = 1)
    ldf.loc[:, "shortName"] = [ShortName(n) for n in ldf.loc[:, "Name"]]
    return ldf
```

```python
def gtz(number):
    if number > 0:
        return 1
    return 0
```

```python
def results(a, b):
    if a > b:
        return ["win", "loss"]
    elif b > a:
        return ["loss", "win"]
    else:
        return ["draw", "draw"]   
```

```python
def simulate_match(matchdf, iterations = 1000):
    df = matchdf.loc[(matchdf["type.name"] == "Shot") & (matchdf["period"] < 5), ["team.name", "shot.statsbomb_xg"]].copy()
    iters = []
    for i in range(iterations):
        df["sim_goals"] = [1 - gtz(random.random() - n) for n in df["shot.statsbomb_xg"].tolist()]
        simmed = pd.DataFrame(df.groupby("team.name")["sim_goals"].sum())
        simmed["results"] = results(simmed.iloc[0, 0], simmed.iloc[1, 0])
        iters.append(simmed)   
    
    sims = pd.concat(iters)
    grouped = (sims.groupby(sims.index)["results"].value_counts()/iterations).unstack()
    if "draw" in grouped.columns and "win" in grouped.columns:
        grouped["points"] = grouped["draw"] +  3 * grouped["win"]
    elif "win" in grouped.columns:
        grouped["points"] = 3 * grouped["win"]
    elif "draw" in grouped.columns:
        grouped["points"] = grouped["draw"]
    
    return grouped
```

```python
def simulate_match1(matchdf, iterations = 1000):
    df = matchdf.loc[(matchdf["type.name"] == "Shot") & (matchdf["period"] < 5), ["team.name", "shot.statsbomb_xg"]].copy()
    iters = []
    for i in range(iterations):
        df["sim_goals"] = [1 - gtz(random.random() - n) for n in df["shot.statsbomb_xg"].tolist()]
        simmed = pd.DataFrame(df.groupby("team.name")["sim_goals"].sum())
        iters.append(simmed)   
    
    sims = pd.concat(iters)
    zips = zip(sims.sim_goals[::2], sims.sim_goals[1::2])
    result = [results(v1, v2) for v1, v2 in zips]
    sims["results"] = [item for sublist in result for item in sublist]
    grouped = sims.groupby(sims.index)["results"].value_counts(normalize = True).unstack()
    
    points = "xPoints"
    if "draw" in grouped.columns and "win" in grouped.columns:
        grouped[points] = grouped["draw"] +  3 * grouped["win"]
    elif "win" in grouped.columns:
        grouped[points] = 3 * grouped["win"]
    elif "draw" in grouped.columns:
        grouped[points] = grouped["draw"]

    return grouped
```

```python
def MatchShotMap(matchdf, overdf):
    df = matchdf

    fig, ax = draw_pitch(pitch_color = "white", line_color = "black", y_size = 8)
    
    overview = overdf.loc[overdf["match_id"] == match_id, :]
    overviewdict = overview.to_dict(orient = "records")[0]
    
    home_team = overviewdict["home_team.home_team_name"]
    away_team = overviewdict["away_team.away_team_name"]
    teams = [home_team, away_team]
    
    cmap = "rainbow"
    goals = []
    shots = []
    xGs = []
    
    scatter = {"size": 100, "alpha": 0.6}
    for team in teams:
        things = {}
        things["shots"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Shot") & (df["shot.outcome.name"] != "Goal") & (df["team.name"] == team), :]
        things["goals"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Shot") & (df["shot.outcome.name"] == "Goal") & (df["team.name"] == team), :]
        things["ogs"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Own Goal Against") & (df["team.name"] != team), :]
        shootout = df.loc[(df["period"] == 5) & (df["type.name"] == "Shot") & (df["shot.outcome.name"] == "Goal") & (df["team.name"] == team), :]
        
        g = str(len(things["goals"]) + len(things["ogs"]))
        if len(shootout) > 0:
            g += " (" + str(len(shootout)) + ")"
        goals.append(g)
        shots.append(len(things["shots"]) + len(things["goals"]))
        xGs.append(things["shots"]["shot.statsbomb_xg"].sum() + things["goals"]["shot.statsbomb_xg"].sum())
        
        for key in ["shots", "goals", "ogs"]:
            thing = things[key]
            if (team == home_team and key in ["shots", "goals"]) or (team == away_team and key in ["ogs"]):
                x = thing["location_x"]
                y = 80 - thing["location_y"]
            else:
                x = 120 - thing["location_x"]
                y = thing["location_y"]
            
            color = thing["shot.statsbomb_xg"]/0.8
            if key == "ogs":
                plt.scatter(x, y, s = scatter["size"], marker = "x", zorder = 2, alpha = scatter["alpha"])
            elif key == "shots":
                plt.scatter(x, y, s = scatter["size"], marker = "o", zorder = 2, alpha = scatter["alpha"], c = color,
                            cmap = cmap, edgecolor = "black", vmin = 0, vmax = 1)
            elif key == "goals":
                plt.scatter(x, y, s = scatter["size"], marker = "s", zorder = 2, alpha = scatter["alpha"], c = color,
                            cmap = cmap, edgecolor = "black", vmin = 0, vmax = 1)
    
    texts = [[round(xGs[0], 2), "xG", round(xGs[1], 2)], [shots[0], "Shots", shots[1]],
            [goals[0], "Score", goals[1]], [home_team, "v", away_team]]
                                                                                               
    start = 83
    inc = 3
    for text in texts:
        if text[0] == home_team: f = 12
        else: f = 10
        plt.text(55, start, text[0], horizontalalignment = "right", fontsize = f, weight = "bold")
        plt.text(60, start, text[1], horizontalalignment = "center", fontsize = f)
        plt.text(65, start, text[2], horizontalalignment = "left", fontsize = f, weight = "bold")
        start += inc

    plt.text(60, start, overviewdict["match_date"], horizontalalignment = "center")   
    plt.text(60, start + inc, overviewdict["competition.competition_name"], horizontalalignment = "center")
    
    plt.tight_layout()
    return fig, ax, plt
    
```

```python
def MatchStep(matchdf, lineupdf, overdf):
    df = matchdf
    
    ldf = lineupdf

    fig, ax = plt.subplots(figsize = (12, 8))
    
    overview = overdf.loc[overdf["match_id"] == match_id, :]
    overviewdict = overview.to_dict(orient = "records")[0]
    
    home_team = overviewdict["home_team.home_team_name"]
    away_team = overviewdict["away_team.away_team_name"]
    teams = [home_team, away_team]

    halves = df.loc[df["type.name"] == "Half End", "time"].unique()
    
    goals = []
    shots = []
    xGs = []
    for i in [0, 1]:
        team = teams[i]
        color = "C" + str(i)
        
        things = {}
        things["shots"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Shot") & (df["team.name"] == team), :].copy()
        things["shots"].loc[:, "xGsum"] = things["shots"]["shot.statsbomb_xg"].cumsum()
        
        things["goals"] = things["shots"].loc[things["shots"]["shot.outcome.name"] == "Goal"]       
        things["ogs"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Own Goal Against") & (df["team.name"] != team), :]
        shootout = df.loc[(df["period"] == 5) & (df["type.name"] == "Shot") & (df["shot.outcome.name"] == "Goal") & (df["team.name"] == team), :]
        
        g = str(len(things["goals"]) + len(things["ogs"]))
        if len(shootout) > 0:
            g += " (" + str(len(shootout)) + ")"
        goals.append(g)
        shots.append(len(things["shots"]))
        xGs.append(things["shots"]["shot.statsbomb_xg"].sum())
        
        x = [0] + things["shots"]["time"].values.tolist() + [halves.max()]
        y = [0] + things["shots"]["xGsum"].values.tolist() + [things["shots"]["xGsum"].max()]
        plt.step(x, y, where = "post", color = color)
        plt.scatter(things["goals"]["time"], things["goals"]["xGsum"], marker = "o", color = color)
        
        for index, row in things["goals"].iterrows():
            name = ldf.loc[ldf["player_id"] == row["player.id"], "shortName"].values.tolist()[0]
            if row["shot.type.name"] == "Penalty":
                name += " (P)"
            plt.annotate(xy = (row["time"] - 0.5 , row["xGsum"] * 1.02), s = name, horizontalalignment = "right")

        for index, row in things["ogs"].iterrows():
            name = ldf.loc[ldf["player_id"] == row["player.id"], "shortName"].values[0] + " (OG)"
            og = things["shots"].loc[things["shots"]["time"] < row["time"], :]
            plt.annotate(xy = (og["time"].max() - 0.5, og["xGsum"].max() * 1.02), s = name, horizontalalignment = "right")
    
    height = max(xGs)
    width = halves.max() / 2
    
    texts = [[round(xGs[0], 2), "xG", round(xGs[1], 2)], [shots[0], "Shots", shots[1]],
            [goals[0], "Score", goals[1]], [home_team, "v", away_team]]
                                                                                               
    start = 1.1
    diff = 0.04
    for text in texts:
        if text[0] == home_team: f, c0, c1 = 12, "C0", "C1"
        else: f, c0, c1 = 10, "k", "k"
        h = start * height
        plt.text(width * 0.9, h, text[0], horizontalalignment = "right", fontsize = f, weight = "bold", color = c0)
        plt.text(width, h, text[1], horizontalalignment = "center", fontsize = f)
        plt.text(width * 1.1, h, text[2], horizontalalignment = "left", fontsize = f, weight = "bold", color = c1)
        start += diff

    plt.text(width, start * height, overviewdict["match_date"], horizontalalignment = "center")
    plt.text(width, (start + diff) * height, overviewdict["competition.competition_name"], horizontalalignment = "center")
        
    for side in ["top", "bottom", "left", "right"]:
        ax.spines[side].set_visible(False)

    for half in halves:
        plt.axvline(half, color = "C2")
    plt.axvline(color = "k")
    plt.axhline(color = "k")
    
    xticks = range(0, int(halves.max()), 15)
    plt.xticks(xticks)
    plt.grid(True)
    
    plt.tight_layout()
    return fig, ax, plt
```

```python
def ratiodf(df):
    df = df.copy()
    df.loc[:, "possession%"] = df.loc[:, "possession"] / df.loc[:, "possessions"]
    df.loc[:, "xG/shot"] =  df.loc[:, "shot.statsbomb_xg"] / df.loc[:, "type.Shot"]
    df.loc[:, "NPxG/shot"] =  df.loc[:, "shot.statsbomb_xg.Non Penalty"] / df.loc[:, "shot.Non Penalty"]
    df.loc[:, "aerialwin%"] = df.loc[:, "aerialwin"] / df.loc[:, df.columns.intersection(["duel.Aerial Lost", "aerialwin"])].sum(axis=1)
    df.loc[:, "dribble%"] = df.loc[:, "dribble.Complete"] / df.loc[:, "type.Dribble"]
    df.loc[:, "NPshot%"] = df.loc[:, df.columns.intersection(["shot.Non Penalty.Goal", "shot.On T.No Goal"])].sum(axis = 1) / df.loc[:, "shot.Non Penalty"]
    df.loc[:, "pass%"] = df.loc[:, "pass.Complete"] / df.loc[:, "type.Pass"]

    return df
```

```python
def teamMatchSummary(matchdf, overdf, ratio = False):
    overdict = overdf.loc[overdf["match_id"] == match_id, :].to_dict(orient = "records")[0]
    teams = {}
    teams[overdict["home_team.home_team_id"]] = overdict["home_team.home_team_name"]
    teams[overdict["away_team.away_team_id"]] = overdict["away_team.away_team_name"]

    df = matchdf.loc[matchdf["period"] < 5, :].copy()

    grouped = df.groupby("team.id")

    conc_list = []
    for t in ["shot.", "pass.", "duel."]:
        gb = ["team.id", t + "outcome.name", t + "type.name"]
        if t == "shot.":
            agg = {"index": "count", "shot.statsbomb_xg": sum}
        else:
            agg = {"index": "count"}
        temp = df.groupby(gb).agg(agg).unstack().unstack()
        temp.columns = [".".join(".".join(n).split(".")[1:]) for n in temp.columns]
        conc_list.append(temp.add_prefix(t))

    for t in ["type.name", "dribble.outcome.name"]:
        pref = t.split(".")[0] + "."
        temp = grouped[t].value_counts().unstack().add_prefix(pref)
        conc_list.append(temp)
        
    tdf = pd.DataFrame(grouped["possession"].nunique())
    tdf["possessions"] = sum(tdf["possession"])
    tdf["time.end"] = df["time"].max()
    tdf["time.start"] = 0    
    tdf.loc[:, "match.played"] = 1
    tdf.loc[:, "time.played"] = tdf.loc[:, "time.end"] - tdf.loc[:, "time.start"]
    
    awcol = [n for n in df.columns if "aerial_won" in n]
    progcol = [n for n in df.columns if "progress" in n]
    sumcol = awcol + progcol
    pdf = grouped[sumcol].sum()
    
    conc_list += [tdf, pdf]
        
    sdf = pd.concat(conc_list, axis = 1).fillna(0)
    sdf.loc[:, "opposition.id"] = [sdf.index[i] for i in [1, 0]]
    sdf.loc[:, "opposition.name"] = [teams[i] for i in sdf.loc[:, "opposition.id"].values.tolist()]
    sdf.loc[:, "team.name"] = [teams[i] for i in sdf.index]
    
    sbcols = [n for n in sdf.columns if "statsbomb_xg" in n]
    sbnpcols = [n for n in sbcols if "Penalty" not in n]
    sbpencols = [n for n in sbcols if "Penalty" in n]

    pencols = [n for n in sdf.columns if "shot" in n and "statsbomb_xg" not in n and "Penalty" in n]
    ngotcols = [n for n in sdf.columns if "shot" in n and "Saved" in n and "statsbomb_xg" not in n and "Penalty" not in n]
    golcol = [n for n in sdf.columns if "shot" in n and "statsbomb_xg" not in n and "Goal" in n]
    
    sdf.loc[:, "shot.statsbomb_xg.Non Penalty"] = sdf.loc[:, sbnpcols].sum(axis = 1)
    sdf.loc[:, "shot.statsbomb_xg"] = sdf.loc[:, sbcols].sum(axis = 1)
    sdf.loc[:, "shot.On T.No Goal"] = sdf.loc[:, ngotcols].sum(axis = 1)
    sdf.loc[:, "shot.Non Penalty"] = sdf.loc[:, "type.Shot"] - sdf.loc[:, pencols].sum(axis = 1)
    sdf.loc[:, "shot.Goal"] = sdf.loc[:, golcol].sum(axis = 1)
    if "shot.Goal" in sdf.columns:
        if "shot.Penalty.Goal" in sdf.columns:
            sdf.loc[:, "shot.Non Penalty.Goal"] = sdf.loc[:, "shot.Goal"] - sdf.loc[:, "shot.Penalty.Goal"]
        else:
            sdf.loc[:, "shot.Non Penalty.Goal"] = sdf.loc[:, "shot.Goal"]
            
    sdf.loc[:, "progress1"] = sdf.loc[:, ["Pass_progress1", "Carry_progress1"]].sum(axis = 1)
    sdf.loc[:, "deepprogress"] = sdf.loc[:, ["Pass_deepprogress", "Carry_deepprogress"]].sum(axis = 1)
    sdf.loc[:, "turnover"] = sdf.loc[:, ["type.Miscontrol", "dribble.Incomplete"]].sum(axis = 1)
    
    sdf.loc[:, "aerialwin"] = sdf.loc[:, sdf.columns.intersection(awcol)].sum(axis = 1)

    passcol = [n for n in sdf.columns if "pass" in n and "x" not in n and "assist" not in n]
    pccol = [n for n in sdf.columns if "Complete" in n and "pass" in n]
    sdf.loc[:, "pass.Complete"] = sdf.loc[:, pccol].sum(axis = 1)

    tacklecol = [n for n in sdf.columns if "duel" in n and "Tackle" in n]
    sdf.loc[:, "duel.Tackle"] = sdf.loc[:, tacklecol].sum(axis = 1)
    for duel in "type.Interception", "duel.Tackle":
        sdf.loc[:, duel + ".PAdj"] = sdf.loc[:, duel] * 2/(1 + np.e ** ( -10 * ((sdf.loc[:, "possession"]/sdf.loc[:, "possessions"]) - 0.5)))
    
    adf = sdf.groupby("opposition.id").sum().add_suffix(".Against")
    sdf = pd.concat([sdf, adf], axis = 1)

    for col in ["match_id", "match_date", "competition.competition_id", "competition.country_name", 
                "competition.competition_name", "season.season_id", "season.season_name"]:
        sdf.loc[:, col] = overdict[col]
    return sdf
```

```python
def columnFlatten(cols, prefix ):
    fixDict = {"index": prefix, "True": "Cross", "False": "Non Cross"}
    new_cols = []
    for col in cols:
        col = [str(c) for c in col]
        for key in fixDict.keys():
            if key in col:
                col[col.index(key)] = fixDict[key]
        new_cols.append(".".join(col))
    return new_cols
        
```

```python
def PossessionAdjust(df, cols):
    for col in cols:
        sigma = 2 / (1 + np.e ** (-10 * ((df.loc[:, "possession"] / df.loc[:, "possessions"]) - 0.5)))
        df.loc[:, col + ".PAdj"] = df.loc[:, col] * sigma
    return df
```

```python
def MatchSummary(matchdf, lineupdf, overdf, player_or_team, ratio = False):
    if player_or_team not in ["player", "team"]:
        return "fail"
    
    if player_or_team == "player":
        grouper = "player.id"
    elif player_or_team == "team":
        grouper = "team.id"
       
    overdict = overdf.loc[overdf["match_id"] == match_id, :].to_dict(orient = "records")[0]
    teams = {}
    teams[overdict["home_team.home_team_id"]] = overdict["home_team.home_team_name"]
    teams[overdict["away_team.away_team_id"]] = overdict["away_team.away_team_name"]

    ldf = lineupdf
    if player_or_team == "player":
        ldf = ldf.rename(columns = {"player_id": "player.id"})
        ldf.set_index("player.id", inplace = True)

    df = matchdf.loc[matchdf["period"] < 5, :].copy()
    
    #xA DataFrame
    for col in ["pass.shot_assist", "pass.goal_assist"]:
        if col not in df.columns:
            df.loc[:, col] = np.NaN
    assists = df.loc[:, "pass.assisted_shot_id"].dropna()
    shots = df.loc[(df["id"]).isin(assists), ["id", "shot.statsbomb_xg"]].rename(columns = {"shot.statsbomb_xg": "pass.xA"})
    df = df.merge(right = shots, left_on = "pass.assisted_shot_id", right_on = "id", how = "outer")

    grouped = df.groupby(grouper)
    
    conc_list = []
    
    for t in ["type.name", "dribble.outcome.name", "shot.type.name", "shot.outcome.name",
              "pass.outcome.name", "duel.type.name"]:
        pref = t.split(".")[0] 
        if pref == "shot":
            temp = df.groupby([grouper, t]).agg({"index": "count", "shot.statsbomb_xg": sum}).unstack()
            temp.columns = columnFlatten(temp.columns, prefix = pref)
        else:
            temp = grouped[t].value_counts().unstack().add_prefix(pref + ".")
        conc_list.append(temp)
    
    for t in [["pass.outcome.name", "pass.cross"], ["shot.outcome.name", "shot.type.name"],
             ["duel.outcome.name", "duel.type.name"]]:
        gb = [grouper] + t
        pref = t[0].split(".")[0]
        if pref == "shot":
            aggFunc = {"index": "count", "shot.statsbomb_xg": sum}
        else:
            aggFunc = {"index": "count"}
        temp = df.groupby(gb).agg(aggFunc).unstack().unstack()
        temp.columns = columnFlatten(temp.columns, prefix = pref)
        conc_list.append(temp) 
    
    #time & possession DataFrame

    match_time = df.time.max()
    tlist = []
    for row in df.loc[df["type.name"] == "Starting XI", ["tactics.lineup", "team.id"]].itertuples():
        lu = pd.json_normalize(row[1])
        lu.loc[:, "team.id"] = row[2]
        tlist.append(lu)

    tdf = pd.concat(tlist)[["player.id", "team.id"]]
    tdf.loc[:, "match.start"] = 1
    subs_off = df.loc[df["type.name"] == "Substitution", ["time", "player.id", "team.id"]].rename(columns={"time": "time.end"})
    subs_on = df.loc[df["type.name"] ==  "Substitution", ["time", "substitution.replacement.id", "team.id"]].rename(columns={"time": "time.start", "substitution.replacement.id": "player.id"})

    tdf = tdf.merge(subs_off, how = "outer")
    tdf = tdf.merge(subs_on, how = "outer")
    tdf.fillna({"time.start": 0, "time.end": match_time, "match.start": 0}, inplace = True)

    if player_or_team == "player":
        tdf["checks"] = list(zip(*[tdf["time.start"], tdf["time.end"]]))

        for check in tdf.checks.unique():
            pdf = df.loc[(df["time"] >= check[0]) & (df["time"] <= check[1]), ["possession_team.id", "possession"]]
            gdf = pdf.groupby("possession_team.id")["possession"].nunique()
            tdf.loc[tdf["checks"] == check, "possessions"] = gdf.sum()
            for team in gdf.index:
                tdf.loc[(tdf["team.id"] == team) & (tdf["checks"] == check), "possession"] = gdf[team]
        tdf.drop(labels = ["team.id", "checks"], axis = 1, inplace = True)
        tdf.set_index(grouper, inplace = True)
        
    else:
        tdf = pd.DataFrame(grouped["possession"].nunique())
        tdf["possessions"] = sum(tdf["possession"])
        tdf["time.end"] = df["time"].max()  
        tdf["time.start"] = 0

    tdf.loc[:, "match.played"] = 1
    tdf.loc[:, "time.played"] = tdf.loc[:, "time.end"] - tdf.loc[:, "time.start"]
    
    #xGchain
    if player_or_team == "player":
        conc_list += [ldf, xGchain(df)]
        
    awcol = [n for n in df.columns if "aerial_won" in n]
    progcol = [n for n in df.columns if "progress" in n]
    passcol = ["pass.xA", "pass.shot_assist", "pass.goal_assist"]
    sumcols = ["pressure_regain", "pass.cross", "shot.statsbomb_xg"] + awcol + progcol + passcol
    pdf = grouped[sumcols].apply(lambda x: x.sum())
    
    conc_list += [tdf, pdf]
    
    sdf = pd.concat(conc_list, axis = 1).fillna(0)
    
    ngotcols = [n for n in sdf.columns if all(i in n for i in ["shot", "Saved"]) and not all(j in n for j in ["statsbomb_xg", "Penalty"])]
    check_cols = ["shot.statsbomb_xg.Penalty", "shot.Penalty", "shot.Goal", "shot.Penalty.Goal"] + passcol + awcol
    for col in check_cols:       
        if col not in sdf.columns:
            sdf.loc[:, col] = 0
    
    sdf.loc[:, "shot.statsbomb_xg.Non Penalty"] = sdf.loc[:, "shot.statsbomb_xg"] - sdf.loc[:, "shot.statsbomb_xg.Penalty"]
    sdf.loc[:, "shot.Non Penalty"] = sdf.loc[:, "type.Shot"] - sdf.loc[:, "shot.Penalty"]
    sdf.loc[:, "shot.Non Penalty.Goal"] = sdf.loc[:, "shot.Goal"] - sdf.loc[:, "shot.Penalty.Goal"]
    sdf.loc[:, "shot.On T.Non Penalty"] = sdf.loc[:, ngotcols].sum(axis = 1)
    
    sdf.loc[:, "progress1"] = sdf.loc[:, ["Pass_progress1", "Carry_progress1"]].sum(axis = 1)
    sdf.loc[:, "deepprogress"] = sdf.loc[:, ["Pass_deepprogress", "Carry_deepprogress"]].sum(axis = 1)
    sdf.loc[:, "turnover"] = sdf.loc[:, ["type.Miscontrol", "dribble.Incomplete"]].sum(axis = 1)
    
    sdf.loc[:, "aerialwin"] = sdf.loc[:, sdf.columns.intersection(awcol)].sum(axis = 1)  
    
    sdf = PossessionAdjust(sdf, ["type.Interception", "duel.Tackle"])
      
    if player_or_team == "team":
        sdf.loc[:, "opposition.id"] = [sdf.index[i] for i in [1, 0]]
        sdf.loc[:, "opposition.name"] = [teams[i] for i in sdf.loc[:, "opposition.id"].values.tolist()]  
        sdf.loc[:, "team.name"] = [teams[i] for i in sdf.index]
        
        adf = sdf.groupby("opposition.id").sum().add_suffix(".Against")
        sdf = pd.concat([sdf, adf], axis = 1)
    else:
        xgb = (sdf.loc[:, "xGchain"] - sdf.loc[:, "shot.statsbomb_xg.Non Penalty"] - sdf.loc[:, "pass.xA"]).values   
        sdf.loc[:, "pass.xGbuildup"] = [max(0, n) for n in xgb]
        
    for col in ["match_id", "match_date", "competition.competition_id", "competition.country_name", 
                "competition.competition_name", "season.season_id", "season.season_name"]:
        sdf.loc[:, col] = overdict[col]  
    return sdf
        
```

```python
def playerMatchSummary(matchdf, lineupdf, overdf, ratio = False):
    overdict = overdf.loc[overdf["match_id"] == match_id, :].to_dict(orient = "records")[0]

    ldf = lineupdf
    ldf.rename(columns = {"player_id": "player.id"}, inplace = True)
    ldf.set_index("player.id", inplace = True)

    df = matchdf.loc[matchdf["period"] < 5, :].copy()

    grouped = df.groupby("player.id")

    conc_list = []
    for t in ["pass.", "shot.", "duel."]:
        gb = ["player.id", t + "outcome.name", t+"type.name"]
        if t == "shot.":
            agg = {"index": "count", "shot.statsbomb_xg": sum}
        else:
            agg = {"index": "count"}
    
            temp = df.groupby(gb).agg(agg).unstack().unstack()
        temp.columns = [".".join(".".join(n).split(".")[1:]) for n in temp.columns]
        conc_list.append(temp.add_prefix(t))

    for t in ["type.name", "dribble.outcome.name", "shot.outcome.name"]:
        pref = t.split(".")[0] + "."
        temp = grouped[t].value_counts().unstack().add_prefix(pref)
        conc_list.append(temp)

    #xA DataFrame
    for col in ["pass.shot_assist", "pass.goal_assist"]:
        if col not in df.columns:
            df.loc[:, col] = np.NaN
    assists = df.loc[(df["pass.shot_assist"] == True)|(df["pass.goal_assist"] == True), ["pass.assisted_shot_id"]]
    for row in assists.itertuples():
        df.loc[row[0], "pass.xA"] = df.loc[df["id"] == row[1], "shot.statsbomb_xg"].values[0]

    agg = {"pass.xA": sum, "pass.shot_assist": sum, "pass.goal_assist": sum}
    xAdf = grouped.agg(agg)

    #time DataFrame

    match_time = df.time.max()
    tlist = []
    for row in df.loc[df["type.name"] == "Starting XI", ["tactics.lineup", "team.id"]].itertuples():
        lu = pd.json_normalize(row[1])
        lu.loc[:, "team.id"] = row[2]
        tlist.append(lu)

    tdf = pd.concat(tlist)[["player.id", "team.id"]]
    tdf.loc[:, "match.start"] = 1
    subs_off = df.loc[df["type.name"] == "Substitution", ["time", "player.id", "team.id"]].rename(columns={"time": "time.end"})
    subs_on = df.loc[df["type.name"] ==  "Substitution", ["time", "substitution.replacement.id", "team.id"]].rename(columns={"time": "time.start", "substitution.replacement.id": "player.id"})

    tdf = tdf.merge(subs_off, how = "outer")
    tdf = tdf.merge(subs_on, how = "outer")
    tdf.fillna({"time.start": 0, "time.end": match_time, "match.start": 0}, inplace = True)

    tdf["checks"] = list(zip(*[tdf["time.start"], tdf["time.end"]]))

    for check in tdf.checks.unique():
        pdf = df.loc[(df["time"] >= check[0]) & (df["time"] <= check[1]), ["possession_team.id", "possession"]]
        gdf = pdf.groupby("possession_team.id")["possession"].nunique()
        tdf.loc[tdf["checks"] == check, "possessions"] = gdf.sum()
        for team in gdf.index:
            tdf.loc[(tdf["team.id"] == team) & (tdf["checks"] == check), "possession"] = gdf[team]

    tdf.loc[:, "match.played"] = 1
    tdf.loc[:, "time.played"] = tdf.loc[:, "time.end"] - tdf.loc[:, "time.start"]

    tdf.set_index("player.id", inplace = True)
    tdf.drop(labels = ["team.id", "checks"], axis = 1, inplace = True)

    #xGchain

    xgcdf = xGchain(df)

    #progress
    awcol = [n for n in df.columns if "aerial_won" in n]
    progcol = [n for n in df.columns if "progress" in n]
    sumcols = ["pressure_regain"] + awcol + progcol
    pdf = grouped[sumcols].apply(lambda x: x.sum())

    conc_list += [pdf, xgcdf, tdf, xAdf, ldf]

    sdf = pd.concat(conc_list, axis = 1).fillna(0)
    
    sbcols = [n for n in sdf.columns if "statsbomb_xg" in n]
    sbnpcols = [n for n in sbcols if "Penalty" not in n]
    sbpencols = [n for n in sbcols if "Penalty" in n]

    pencols = [n for n in sdf.columns if "shot" in n and "statsbomb_xg" not in n and "Penalty" in n]
    ngotcols = [n for n in sdf.columns if "shot" in n and "Saved" in n and "statsbomb_xg" not in n and "Penalty" not in n]
    golcol = [n for n in sdf.columns if "shot" in n and "statsbomb_xg" not in n and "Goal" in n]
    
    [n for n in sdf.columns if all(i in n for i in ["shot", "Saved"]) and not all(j in n for j in ["statsbomb_xg", "Penalty"])]
    
    sdf.loc[:, "shot.statsbomb_xg.Non Penalty"] = sdf.loc[:, sbnpcols].sum(axis = 1)
    sdf.loc[:, "shot.statsbomb_xg"] = sdf.loc[:, sbcols].sum(axis = 1)
    sdf.loc[:, "shot.On T.No Goal"] = sdf.loc[:, ngotcols].sum(axis = 1)
    sdf.loc[:, "shot.Non Penalty"] = sdf.loc[:, "type.Shot"] - sdf.loc[:, pencols].sum(axis = 1)
    sdf.loc[:, "shot.Goal"] = sdf.loc[:, golcol].sum(axis = 1)
    if "shot.Goal" in sdf.columns:
        if "shot.Penalty.Goal" in sdf.columns:
            sdf.loc[:, "shot.Non Penalty.Goal"] = sdf.loc[:, "shot.Goal"] - sdf.loc[:, "shot.Penalty.Goal"]
        else:
            sdf.loc[:, "shot.Non Penalty.Goal"] = sdf.loc[:, "shot.Goal"]
    if "pass.xA" in sdf.columns:
        sdf.loc[:, "pass.xGbuildup"] = sdf.loc[:, "xGchain"] - sdf.loc[:, "shot.statsbomb_xg.Non Penalty"] - sdf.loc[:, "pass.xA"]
    else:
        sdf.loc[:, "pass.xGbuildup"] = sdf.loc[:, "xGchain"] - sdf.loc[:, "shot.statsbomb_xg.Non Penalty"]
    sdf.loc[:, "progress1"] = sdf.loc[:, ["Pass_progress1", "Carry_progress1"]].sum(axis = 1)
    sdf.loc[:, "deepprogress"] = sdf.loc[:, ["Pass_deepprogress", "Carry_deepprogress"]].sum(axis = 1)
    sdf.loc[:, "turnover"] = sdf.loc[:, ["type.Miscontrol", "dribble.Incomplete"]].sum(axis = 1)
    
    sdf.loc[:, "aerialwin"] = sdf.loc[:, sdf.columns.intersection(awcol)].sum(axis = 1)

    passcol = [n for n in sdf.columns if "pass" in n and "x" not in n and "assist" not in n]
    pccol = [n for n in sdf.columns if "Complete" in n and "pass" in n]
    sdf.loc[:, "pass.Complete"] = sdf.loc[:, pccol].sum(axis = 1)

    tacklecol = [n for n in sdf.columns if "duel" in n and "Tackle" in n]
    sdf.loc[:, "duel.Tackle"] = sdf.loc[:, tacklecol].sum(axis = 1)
    for duel in "type.Interception", "duel.Tackle":
        sdf.loc[:, duel + ".PAdj"] = sdf.loc[:, duel] * 2/(1 + np.e ** ( -10 * ((sdf.loc[:, "possession"]/sdf.loc[:, "possessions"]) - 0.5)))
        
    for col in ["match_id", "match_date", "competition.competition_id", "competition.country_name", 
                "competition.competition_name", "season.season_id", "season.season_name"]:
        sdf.loc[:, col] = overdict[col]
    
    if ratio:
        rdf = ratiodf(sdf)
    else: rdf = sdf
    return rdf
```

```python
def per90(df):
    spare = ["Name", "team.name", "competition.competition_name", "season.season_name", "time.played", 
             "time.start", "time.end"]
    cols = [n for n in df.columns if n not in spare]
    for col in cols:
        df[col] = df[col] * 90 / (df["time.played"])
    return df
```

```python
def Bar(df, stats, bar_style, stat_type = "", minutes = 0):
    bardf = df.copy()
    
    fig, ax = plt.subplots(figsize = (9, 12))
 
    for side in ["top", "bottom",  "right"]:
        ax.spines[side].set_visible(False) 
        
    if stat_type == "per90":
        for stat in stats:
            bardf.loc[:, "{} {}".format(stat, stat_type)] = bardf.loc[:, stat] * 90 / bardf.loc[:, "time.played"]
    elif stat_type == "perMatch":
        for stat in stats:
            bardf.loc[:, "{} {}".format(stat, stat_type)] = bardf.loc[:, stat]/bardf.loc[:, "match.played"]
            
    stats = [("{} {}".format(n, stat_type)).strip() for n in stats]
    if bar_style == "plus":
        bardf.loc[:, "stats"] = bardf.loc[:, stats].sum(axis = 1)
        bardf.loc[:, "zeroes"] = 0
    elif bar_style == "minus":
        bardf.loc[:, "stats"] = bardf.loc[:, stats[0]] - bardf.loc[:, stats[1]]
        
    gdf = bardf.loc[bardf["time.played"] > minutes, :]
    
    cvalue = min([30, len(gdf)])
    cdf = gdf.sort_values("stats", ascending = False)[:cvalue]
    
    tests, sub, lab = {}, [],  []
    cols = ["Name", "team.name", "competition.competition_name", "season.season_name"]
    for col in cols:
        if col in cdf.columns:
            tests[col] = cdf.loc[:, col].unique()
            if len(tests[col]) == 1:
                sub.append(tests[col][0])
            else:
                lab.append(col)
         
    if len(lab) == 0:
        cdf["labels"] = list(zip(*[cdf[n] for n in cols if n not in ["Name", "team.name"]]))
    else:
        cdf["labels"] = list(zip(*[cdf[n] for n in lab]))
      
    y = np.arange(len(cdf), 0, -1)
    if bar_style == "plus":
        x = [cdf.loc[:, "zeroes"]] + [cdf.loc[:, stat ] for stat in stats]
        for i in range(len(stats)):
            plt.barh(y = y, width = x[i+1], height = 0.5, left = sum(x[:(i+1)]), 
                     tick_label = cdf["labels"], label = stats[i])  
    elif bar_style == "minus":
        x1 = cdf.loc[:, stats[0]]
        x2 = cdf.loc[:, stats[1]]        
        
        adj = 0.2
        plt.barh(y = y + adj, width = x1, height = adj * 2 - 0.05,
                     tick_label = cdf["labels"], label = stats[0])
        plt.barh(y = y - adj, width = x2, height = adj * 2 - 0.05,
                     tick_label = cdf["labels"], label = stats[1])
        

    x, new_y = cdf.stats.max()/2, max(y) + 1
    if minutes > 0:
        new_y += 0.5
        plt.text(x = x, y = new_y, s = "Minimum: {} minutes played".format(minutes), horizontalalignment = "center")
    if len(sub) > 0:
        new_y += 0.5
        plt.text(x = x, y = new_y, s = ", ".join(sub), horizontalalignment = "center")

    title = " {} ".format(bar_style)
    plt.text(x = x, y = new_y + 0.5 , s = title.join(stats), 
             horizontalalignment = "center", fontsize = 12)
    
    plt.grid(True)
    ax.legend(loc = "lower right")
    plt.tight_layout()
    
    return fig, ax, plt
```

```python
def Scatter(df, stats, quintile, stat_type = "", minutes = 0):
    fig, ax = plt.subplots(figsize = (10, 10)) 
    
    scatterdf = df.copy()

    for side in ["top", "bottom",  "right", "left"]:
        ax.spines[side].set_visible(False) 
            
    for stat in stats:
        if any([n in stat for n in ["%", "/"]]):
            scatterdf.loc[:, "{} {}".format(stat, stat_type)] = scatterdf.loc[:, stat]
        else:
            if stat_type == "per90":
                scatterdf.loc[:, "{} {}".format(stat, stat_type)] = scatterdf.loc[:, stat] * 90 / scatterdf.loc[:, "time.played"]
            elif stat_type == "perMatch":
                scatterdf.loc[:, "{} {}".format(stat, stat_type)] = scatterdf.loc[:, stat]/scatterdf.loc[:, "match.played"]
            
            
    stats = [("{} {}".format(n, stat_type)).strip() for n in stats]
    scatterdf["stats"] = scatterdf[stats[0]]/scatterdf[stats[0]].max() + scatterdf[stats[1]]/scatterdf[stats[1]].max()

    gdf = scatterdf.loc[df["time.played"] >= minutes].copy()
    
    tests, sub, lab = {}, [],  []  
    cols = ["team.name", "competition.competition_name", "season.season_name"]
    if pot == "player":
        cols = ["Name"] + cols
    for col in cols:
        tests[col] = gdf.loc[:, col].unique()
        if len(tests[col]) == 1:
            sub.append(tests[col][0])
        else:
            if col not in ["Name", "team.name"]:
                lab.append(col)

    if len(lab) == 0:
        gdf["labels"] = list(zip(*[gdf[n] for n in cols if n not in ["Name", "team.name"]]))
    else:
        gdf["labels"] = list(zip(*[gdf[n] for n in lab]))

    for label in gdf["labels"].unique():
        x = gdf.loc[gdf["labels"] == label, stats[0]]
        y = gdf.loc[gdf["labels"] == label, stats[1]]
        plt.scatter(x = x, y = y, s = 30, label = label, edgecolor = "k", alpha = 0.8, linewidth = 0.5)
    
    q = quintile
    x = gdf.loc[:, stats[0]]
    y = gdf.loc[:, stats[1]]

    adf = gdf.loc[(gdf[stats[0]] > x.quantile(q))|(gdf[stats[1]] > y.quantile(q))|(gdf["stats"] > gdf["stats"].quantile(q)), :]
    if "Name" in scatterdf.columns:
        names = adf.loc[:, "Name"].apply(ShortName)
    else:
        names = adf.loc[:, "team.name"]
    for i, name in enumerate(names):
        xa = adf.loc[:, stats[0]].values[i] + x.max()/100
        ya = adf.loc[:, stats[1]].values[i] - y.max()/60
        plt.annotate(s = name, xy = (xa, ya))
    
    plt.axvline(c = "k")
    plt.axhline(c = "k")
        
    plt.xlabel(stats[0])
    plt.ylabel(stats[1])
    
    x, new_y, elev = x.max()/2, max(y) * 1.1, max(y)/25
    if minutes > 0:
        new_y += elev
        plt.text(x = x, y = new_y, s = "Minimum: {} minutes played".format(minutes), horizontalalignment = "center")
    if len(sub) > 0:
        new_y += elev
        plt.text(x = x, y = new_y, s = ", ".join(sub), horizontalalignment = "center")

    plt.text(x = x, y = new_y + elev , s = " against ".join(stats), horizontalalignment = "center", fontsize = 12)
    
    plt.grid(True)
    ax.legend(loc = "upper right")
    plt.tight_layout()
    
    return fig, ax, plt
```
