```python
def timestamptoseconds(timestamp):
    broken = timestamp.split(":")
    s = sum([float(broken[i]) * (60 ** (2 - i)) for i in range(len(broken))])
    return s

def timestamptominutes(timestamp):
    m = timestamptoseconds(timestamp) / 60
    return m    

def tts(timestamps):
    ts = [timestamptoseconds(n) for n in timestamps]
    return ts

def ttm(timestamps):
    ts = [timestamptominutes(n) for n in timestamps]
    return ts
```

```python
def positivenegative(a):
    if a > 0:
        return 1
    elif a < 0:
        return -1
    else:
        return 0
```

```python
def json_loads(filepath):
    with open(filepath, encoding="utf-8") as json_file:
        return json.load(json_file)
```

```python
def locationFinder(row):
    end_cols = [n for n in row.index if "end_location" in n]
    for col in end_cols:
        if isinstance(row[col], list):
            return row[col]
    return row[col]
```

```python
def locationFinder1(row):
    xs = row.values
    for x in xs:
        if isinstance(x, list):
            return x
```

```python
def progress1(row):
    if (row["location_x"] > 60) and (row["end_location_x"] - row["location_x"] > 10) and (row["type.name"] in ["Pass", "Carry"]):
        return 1
    return np.NaN
```

```python
def progress2(row):
    if row["location_x"] > 60:
        if row["end_location_x"] - row["location_x"] > 0:
            return row["end_location_x"] - row["location_x"]
    return np.NaN
```

```python
def deepprogress(row):
    if (row["location_x"] <= 80) and (row["end_location_x"] > 80) and (row["type.name"] in ["Pass", "Carry"]):
        return 1
    return np.NaN
```

```python
def json_dataFrame(filepath):
    json_file = json_loads(filepath)
    df = pd.json_normalize(json_file)
    
    check_columns = {"location", "period", "timestamp"}
    if check_columns.issubset(df.columns): 
        x = df.groupby(["period"])["timestamp"].last()
        x[0] = "00:00:00"
        vfunc = np.vectorize(timestamptominutes)
        y = vfunc(x.sort_index().values).cumsum()
        for i in x.index[:(-1)]:
            df.loc[df["period"] == i, "time"] = vfunc(df.loc[df["period"] == i, "timestamp"]) + y[i-1]
        
        df["end_location"] = df[[n for n in df.columns if "end_location" in n]].dropna(thresh = 1).apply(locationFinder1, axis = 1)
        
        suffs = ["_x", "_y"]
        for pre in ["end_", ""]:
            for i in range(len(suffs)):
                suff = suffs[i]
                df.loc[:, "{}location{}".format(pre, suff)] = [n[i] if isinstance(n, list) else np.NaN for n in df[pre + "location"] ]
        
        #Pressures
        pdf = df.loc[df["type.name"] == "Pressure", :]
        for i, row in pdf.iterrows():
            t = row["time"] + 1/12
            rdf = df.loc[(df["time"] >= row["time"]) & (df["time"] <= t), "possession"].max()
            if rdf > row["possession"]:
                df.loc[i, "pressure_regain"] = 1
        
        #Progress
        progs = ["Pass", "Carry"]
        for prog in progs:  
            df.loc[(df["type.name"] == prog), prog + "_progress1"] = df.apply(progress1, axis = 1)
            #df.loc[(df["type.name"] == prog), prog + "_progress2"] = df.apply(progress2, axis = 1)
        df.loc[df["type.name"].isin(progs), "deepprogress"] = df.apply(deepprogress, axis = 1)
        
    return df
```

```python
def NameColumn(row):
    if row["player_nickname"] == None:
        name = row["player_name"]
    else:
        name = row["player_nickname"]
    return name.strip()
```

```python
def ShortName(name):
    split = name.split(" ")
    if len(split) == 1:
        return name
    else:
        return split[0][0] + ". " + split[-1]
```

```python
def draw_pitch(pitch_color = "green", line_color = "white", y_size = 10):
    x_size = y_size * 1.5
    fig, ax = plt.subplots(figsize = (x_size, y_size))
    ax.axis("off")
    fig.patch.set_facecolor(pitch_color)
    for side in ["top", "bottom", "left", "right"]:
        ax.spines[side].set_visible(False)   
    
    x1, x2 = -5, 125
    y1, y2 = -5, 85
    z = 1
    
    plt.xlim(x1, x2)
    plt.ylim(y1, y2)
    
    Pitch = pch.Rectangle([0, 0], width = 120, height = 80, fill = False, color = line_color, zorder = z)
    
    Halfway = pch.ConnectionPatch([60, 0], [60, 80], "data", "data", color = line_color, zorder = z)
    CentreCircle = pch.Circle((60, 40), 10, color = line_color, fill = False, zorder = z)
    CentreSpot = pch.Circle((60, 40), 0.3, color = line_color, fill = True, zorder = z)
    
    LeftPenalty = pch.Rectangle([0, 18], width = 18, height = 44, color = line_color, fill = False, zorder = z)
    RightPenalty = pch.Rectangle([102, 18], width = 18, height = 44, color = line_color, fill = False, zorder = z)
    
    LeftPenSpot = pch.Circle((12, 40), 0.3, color = line_color, fill = True, zorder = z)
    RightPenSpot = pch.Circle((108, 40), 0.3, color = line_color, fill = True, zorder = z)
    
    LeftSixYard = pch.Rectangle([0, 30], width = 6, height = 20, color = line_color, fill = False, zorder = z)
    RightSixYard = pch.Rectangle([114, 30], width = 6, height = 20, color = line_color, fill = False, zorder = z)
    
    LeftGoal = pch.Rectangle([-3, 36], width = 3, height = 8, color = line_color, fill = False, zorder = z)
    RightGoal = pch.Rectangle([120, 36], width = 3, height = 8, color = line_color, fill = False, zorder = z)
    
    LeftArc = pch.Arc((12, 40), height = 17.8, width = 17.8, angle = 0, theta1 = 310, theta2 = 50, color = line_color, zorder = z)
    RightArc = pch.Arc((108, 40), height = 17.8, width = 17.8, angle = 0, theta1 = 130, theta2 = 230, color = line_color, zorder = z)
    
    for i in [Pitch, Halfway, CentreCircle, CentreSpot,
              LeftPenalty, RightPenalty, LeftPenSpot, RightPenSpot,
              LeftSixYard, RightSixYard, LeftGoal, RightGoal, LeftArc, RightArc]:
        ax.add_patch(i)
        
    return fig, ax, plt
```

```python
def LineupDataFrame(match_id):
    lpath = os.path.join(lineups, str(match_id) + ".json")
    lineup = json_loads(lpath)
    llist = []
    for i in range(len(lineup)):
        df = pd.json_normalize(lineup[i]["lineup"])
        df.loc[:, "team.name"] = lineup[i]["team_name"]
        df.loc[:, "team.id"] = lineup[i]["team_id"]
        llist.append(df)
    ldf = pd.concat(llist)

    ldf.loc[:, "Name"] = ldf.apply(NameColumn, axis = 1)
    ldf.loc[:, "shortName"] = ldf["Name"].apply(ShortName)
    return ldf
```

```python
def MatchShotMap(match_id, overdf):
    path = os.path.join(events, str(match_id) + ".json")
    df = json_dataFrame(path)

    fig, ax, plt = draw_pitch(pitch_color = "white", line_color = "black", y_size = 8)
    
    overview = overdf.loc[overdf["match_id"] == match_id, :]
    overviewdict = overview.to_dict(orient = "records")[0]
    
    home_team = overviewdict["home_team.home_team_name"]
    away_team = overviewdict["away_team.away_team_name"]
    teams = [home_team, away_team]
    
    cmap = "rainbow"
    goals = []
    shots = []
    xGs = []
    
    scatter = {"size": 100, "alpha": 0.6}
    for team in teams:
        things = {}
        things["shots"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Shot") & (df["shot.outcome.name"] != "Goal") & (df["team.name"] == team), :]
        things["goals"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Shot") & (df["shot.outcome.name"] == "Goal") & (df["team.name"] == team), :]
        things["ogs"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Own Goal Against") & (df["team.name"] != team), :]
        shootout = df.loc[(df["period"] == 5) & (df["type.name"] == "Shot") & (df["shot.outcome.name"] == "Goal") & (df["team.name"] == team), :]
        
        g = str(len(things["goals"]) + len(things["ogs"]))
        if len(shootout) > 0:
            g += " (" + str(len(shootout)) + ")"
        goals.append(g)
        shots.append(len(things["shots"]) + len(things["goals"]))
        xGs.append(things["shots"]["shot.statsbomb_xg"].sum() + things["goals"]["shot.statsbomb_xg"].sum())
        
        for key in ["shots", "goals", "ogs"]:
            thing = things[key]
            if (team == home_team and key in ["shots", "goals"]) or (team == away_team and key in ["ogs"]):
                x = thing["location_x"]
                y = 80 - thing["location_y"]
            else:
                x = 120 - thing["location_x"]
                y = thing["location_y"]
            
            color = thing["shot.statsbomb_xg"]/0.8
            if key == "ogs":
                plt.scatter(x, y, s = scatter["size"], marker = "x", zorder = 2, alpha = scatter["alpha"])
            elif key == "shots":
                plt.scatter(x, y, s = scatter["size"], marker = "o", zorder = 2, alpha = scatter["alpha"], c = color,
                            cmap = cmap, edgecolor = "black", vmin = 0, vmax = 1)
            elif key == "goals":
                plt.scatter(x, y, s = scatter["size"], marker = "s", zorder = 2, alpha = scatter["alpha"], c = color,
                            cmap = cmap, edgecolor = "black", vmin = 0, vmax = 1)
    
    texts = [[round(xGs[0], 2), "xG", round(xGs[1], 2)], [shots[0], "Shots", shots[1]],
            [goals[0], "Score", goals[1]], [home_team, "v", away_team]]
                                                                                               
    start = 83
    inc = 3
    for text in texts:
        if text[0] == home_team: f = 12
        else: f = 10
        plt.text(55, start, text[0], horizontalalignment = "right", fontsize = f, weight = "bold")
        plt.text(60, start, text[1], horizontalalignment = "center", fontsize = f)
        plt.text(65, start, text[2], horizontalalignment = "left", fontsize = f, weight = "bold")
        start += inc

    plt.text(60, start, overviewdict["match_date"], horizontalalignment = "center")   
    plt.text(60, start + inc, overviewdict["competition.competition_name"], horizontalalignment = "center")
    
    plt.tight_layout()
    return fig, ax, plt
    
```

```python
def MatchStep(match_id, overdf):
    path = os.path.join(events, str(match_id) + ".json")
    df = json_dataFrame(path)
    
    ldf = LineupDataFrame(match_id)

    fig, ax = plt.subplots(figsize = (12, 8))
    
    overview = overdf.loc[overdf["match_id"] == match_id, :]
    overviewdict = overview.to_dict(orient = "records")[0]
    
    home_team = overviewdict["home_team.home_team_name"]
    away_team = overviewdict["away_team.away_team_name"]
    teams = [home_team, away_team]

    halves = df.loc[df["type.name"] == "Half End", "time"].unique()
    
    goals = []
    shots = []
    xGs = []
    for i in [0, 1]:
        team = teams[i]
        color = "C" + str(i)
        
        things = {}
        things["shots"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Shot") & (df["team.name"] == team), :].copy()
        things["shots"].loc[:, "xGsum"] = things["shots"]["shot.statsbomb_xg"].cumsum()
        
        things["goals"] = things["shots"].loc[things["shots"]["shot.outcome.name"] == "Goal"]       
        things["ogs"] = df.loc[(df["period"] < 5) & (df["type.name"] == "Own Goal Against") & (df["team.name"] != team), :]
        shootout = df.loc[(df["period"] == 5) & (df["type.name"] == "Shot") & (df["shot.outcome.name"] == "Goal") & (df["team.name"] == team), :]
        
        g = str(len(things["goals"]) + len(things["ogs"]))
        if len(shootout) > 0:
            g += " (" + str(len(shootout)) + ")"
        goals.append(g)
        shots.append(len(things["shots"]))
        xGs.append(things["shots"]["shot.statsbomb_xg"].sum())
        
        x = [0] + things["shots"]["time"].values.tolist() + [halves.max()]
        y = [0] + things["shots"]["xGsum"].values.tolist() + [things["shots"]["xGsum"].max()]
        plt.step(x, y, where = "post", color = color)
        plt.scatter(things["goals"]["time"], things["goals"]["xGsum"], marker = "o", color = color)
        
        for index, row in things["goals"].iterrows():
            name = ldf.loc[ldf["player_id"] == row["player.id"], "shortName"].values.tolist()[0]
            if row["shot.type.name"] == "Penalty":
                name += " (P)"
            plt.annotate(xy = (row["time"] - 0.5 , row["xGsum"] * 1.02), s = name, horizontalalignment = "right")

        for index, row in things["ogs"].iterrows():
            name = ldf.loc[ldf["player_id"] == row["player.id"], "shortName"].values[0] + " (OG)"
            og = things["shots"].loc[things["shots"]["time"] < row["time"], :]
            plt.annotate(xy = (og["time"].max() - 0.5, og["xGsum"].max() * 1.02), s = name, horizontalalignment = "right")
    
    height = max(xGs)
    width = halves.max() / 2
    
    texts = [[round(xGs[0], 2), "xG", round(xGs[1], 2)], [shots[0], "Shots", shots[1]],
            [goals[0], "Score", goals[1]], [home_team, "v", away_team]]
                                                                                               
    start = 1.1
    diff = 0.04
    for text in texts:
        if text[0] == home_team: f, c0, c1 = 12, "C0", "C1"
        else: f, c0, c1 = 10, "k", "k"
        h = start * height
        plt.text(width * 0.9, h, text[0], horizontalalignment = "right", fontsize = f, weight = "bold", color = c0)
        plt.text(width, h, text[1], horizontalalignment = "center", fontsize = f)
        plt.text(width * 1.1, h, text[2], horizontalalignment = "left", fontsize = f, weight = "bold", color = c1)
        start += diff

    plt.text(width, start * height, overviewdict["match_date"], horizontalalignment = "center")
    plt.text(width, (start + diff) * height, overviewdict["competition.competition_name"], horizontalalignment = "center")
        
    for side in ["top", "bottom", "left", "right"]:
        ax.spines[side].set_visible(False)

    for half in halves:
        plt.axvline(half, color = "C2")
    plt.axvline(color = "k")
    plt.axhline(color = "k")
    
    xticks = range(0, int(halves.max()), 15)
    plt.xticks(xticks)
    plt.grid(True)
    
    plt.tight_layout()
    return fig, ax, plt
```

```python
def MatchPassMap(team_name, match_id, overdf):
    path = os.path.join(events, str(match_id) + ".json")   
    df = json_dataFrame(path)

    fig, ax, plt = draw_pitch(pitch_color = "white", line_color = "black", y_size = 8)
    
    overview = overdf.loc[overdf["match_id"] == match_id, :]
    overviewdict = overview.to_dict(orient = "records")[0]
    
    home_team = overviewdict["home_team.home_team_name"]
    away_team = overviewdict["away_team.away_team_name"]
    teams = [home_team, away_team]
    
    ldf = LineupDataFrame(match_id)
    
    cmap = cm.get_cmap("rainbow")

    brdf = df.loc[(df["team.name"] == team_name) & (df["type.name"] == "Ball Receipt*"), : ]
    pdf = df.loc[(df["team.name"] == team_name)  & (df["type.name"] == "Pass") & (df["pass.outcome.name"] != "Incomplete"), : ]

    location = brdf.groupby("player.name")[["location_x", "location_y"]].median()
    size = pdf.groupby("player.name")["id"].count()

    newdf = pd.concat([location, size], axis =1 )
       
    if team_name == home_team:
        newdf["x"] = newdf["location_x"]
        newdf["y"] = 80 - newdf["location_y"]
    else:
        newdf["x"] = 120 - newdf["location_x"]
        newdf["y"] = newdf["location_y"]
    plt.scatter(newdf["x"], newdf["y"], s = 15 * newdf["id"], c = "white", edgecolor = "k", zorder = 2)
    
    for i, row in newdf.iterrows():
        name = ldf.loc[ldf["player_name"] == i, "shortName"].values[0] 
        plt.annotate(xy = (row[["x", "y"]] - np.array([0, 4])), s = name, horizontalalignment = "center", zorder = 3)

    pairings = pdf.groupby(["player.name", "pass.recipient.name"])["id"].count().reset_index()
    for i, row in pairings.iterrows():
        start = newdf.loc[row["player.name"], ["x", "y"]]
        end = newdf.loc[row["pass.recipient.name"], ["x", "y"]]
        diff = end - start
        angle = (math.atan2(diff[0], diff[1]) + 1.5 * math.pi) % (2 * math.pi)
        dz = 0.5 * np.array([math.cos(angle), math.sin(angle)])
        
        alpha = min(1, row["id"] / 30)
        plt.arrow(start["x"] + dz[0], start["y"] + dz[1], diff["x"], diff["y"], linewidth = 3, 
                  alpha = alpha, zorder = 1, shape = "left", color = cmap(row["id"]/30))
        
    texts = [overviewdict["home_score"], "Score", overviewdict["away_score"]], [home_team, "v", away_team]
    start = 83
    inc = 3
    for text in texts:
        if text[0] == home_team: f = 12
        else: f = 10
        if text[0] == team_name: color = ["C0", "k", "k"]
        elif text[2] ==  team_name: color = ["k", "k", "C1"]
        else: color = 3 * ["k"]
        plt.text(55, start, text[0], horizontalalignment = "right", fontsize = f, weight = "bold", color = color[0])
        plt.text(60, start, text[1], horizontalalignment = "center", fontsize = f, color = color[1])
        plt.text(65, start, text[2], horizontalalignment = "left", fontsize = f, weight = "bold", color = color[2])
        start += inc

    plt.text(60, start, overviewdict["match_date"], horizontalalignment = "center")   
    plt.text(60, start + inc, overviewdict["competition.competition_name"], horizontalalignment = "center")
    
    plt.tight_layout()
        
    return fig, ax, plt
    
```

```python
def xGchain(df):
    pname = "player.id"
    noPen = df.loc[df["shot.type.name"] != "Penalty", :]
    a =  noPen.groupby(["possession_team.name", "possession", "team.name", pname])["id"].count().reset_index()
    b = noPen.groupby(["possession"])["shot.statsbomb_xg"].max()
    c = a.merge(b, on = "possession", how = "outer")
    d = c.loc[(c["possession_team.name"] == c["team.name"]) & (c["shot.statsbomb_xg"] > 0), :]
    rdf = pd.DataFrame(d.groupby(pname)["shot.statsbomb_xg"].sum())
    rdf.rename(columns = {"shot.statsbomb_xg": "xGchain"}, inplace = True)
    return rdf
```

```python
def playerMatchSummary(match_id, overdf):
    path = os.path.join(events, str(match_id) + ".json")
    matchdf = json_dataFrame(path)
    ldf = LineupDataFrame(match_id)
    ldf.loc[:, "player.id"] = ldf.loc[:, "player_id"]

    df = matchdf.loc[matchdf["period"] < 5, :].copy()
    
    grouped = df.groupby("player.id")
    #Counting
    count_list = []
    list_to_count = ["type.name", "shot.outcome.name", "pass.outcome.name", "dribble.outcome.name", "duel.type.name"]
    for tc in list_to_count:
        prefix = tc.split(".")[0] + "."
        temp = grouped[tc].value_counts().unstack().add_prefix(prefix)
        count_list.append(temp)
        
    temp = df.groupby(["player.id", "shot.type.name", "shot.outcome.name"])["id"].count().reset_index()
    temp = temp.loc[(temp["shot.outcome.name"] == "Goal")&(temp["shot.type.name"] ==  "Penalty"), ["player.id", "id"]]
    temp.rename(columns = {"id": "shot.penalty.goal"}, inplace = True)
    temp.set_index("player.id", inplace = True)
    count_list.append(temp)
    codf = pd.concat(count_list, axis = 1)

    #xA DataFrame
    for col in ["pass.shot_assist", "pass.goal_assist"]:
        if col not in df.columns:
            df.loc[:, col] = np.NaN
    assists = df.loc[(df["pass.shot_assist"] == True)|(df["pass.goal_assist"] == True), :]
    for i, row in assists.iterrows():
        df.loc[i, "pass.xA"] = df.loc[df["id"] == row["pass.assisted_shot_id"], "shot.statsbomb_xg"].values[0]

    agg = {"pass.xA": sum, "pass.shot_assist": sum, "pass.goal_assist": sum}
    xAdf = grouped.agg(agg)
    
    #xG DataFrame
    xG_list = []
    agg = {"shot.statsbomb_xg": [sum, "count"]}

    shots = [df, df.loc[df["shot.type.name"] != "Penalty", :]]
    for i in range(len(shots)):
        cf = shots[i]
        temp = cf.groupby("player.id").agg(agg)
        if i == 1:
             cols = ["shot.NPxG", "shot.NPcount"]
        else:
             cols = ["shot.xG", "shot.count"]
        temp.columns = cols
        xG_list.append(temp)
    xGdf = pd.concat(xG_list, axis = 1)  

    #time DataFrame
    
    match_time = df.time.max()
    tlist = []
    for i, row in df.loc[df["type.name"] == "Starting XI", ["tactics.lineup", "team.id"]].iterrows():
        r = row["tactics.lineup"]
        lu = pd.json_normalize(r)
        lu.loc[:, "team.id"] = row["team.id"]
        tlist.append(lu)

    tdf = pd.concat(tlist)[["player.id", "team.id"]]
    tdf["starter"] = 1
    subs_off = df.loc[df["type.name"] == "Substitution", ["time", "player.id", "team.id"]].rename(columns={"time": "time.end"})
    subs_on = df.loc[df["type.name"] ==  "Substitution", ["time", "substitution.replacement.id", "team.id"]].rename(columns={"time": "time.start", "substitution.replacement.id": "player.id"})

    tdf = tdf.merge(subs_off, how = "outer")
    tdf = tdf.merge(subs_on, how = "outer")
    tdf.fillna({"time.start": 0, "time.end": match_time, "starter": 0}, inplace = True)

    tdf["checks"] = list(zip(*[tdf["time.start"], tdf["time.end"]]))

    for check in tdf.checks.unique():
        pdf = df.loc[(df["time"] >= check[0]) & (df["time"] <= check[1]), ["possession_team.id", "possession"]]
        gdf = pdf.groupby("possession_team.id")["possession"].nunique()
        tdf.loc[tdf["checks"] == check, "possessions"] = gdf.sum()
        for team in gdf.index:
            tdf.loc[(tdf["team.id"] == team) & (tdf["checks"] == check), "possession"] = gdf[team]
            
    tdf.loc[:, "games"] = 1
    tdf.loc[:, "time.played"] = tdf.loc[:, "time.end"] - tdf.loc[:, "time.start"]

    tdf.set_index("player.id", inplace = True)
    tdf.drop(labels = ["team.id", "checks"], axis = 1, inplace = True)
    
    #xGchain
    
    xgcdf = xGchain(df)
    
    #progress
    
    pdf = grouped[["Pass_progress1", "Carry_progress1", "pressure_regain", "deepprogress"]].sum()
    
    #Merging
    new_df = ldf
    for cf in [codf, tdf, xGdf, xAdf, xgcdf, pdf]:
        new_df = new_df.merge(cf, how = "outer", on = "player.id")

    new_df.loc[:, "xGbuildup"] = new_df.loc[:, "xGchain"] - new_df.loc[:, "shot.NPxG"] - new_df.loc[:, "pass.xA"]
    
    for team in new_df["team.id"].unique():
        new_df.loc[new_df["team.id"] != team, "opponent.id"] = team
        new_df.loc[new_df["team.id"] != team, "opponent.name"] = new_df.loc[new_df["team.id"] == team, "team.name"].values[0]
        
    for col in ["match_id", "match_date", "competition.competition_id", "competition.country_name", 
                "competition.competition_name", "season.season_id", "season.season_name"]:
        new_df.loc[:, col] = overdf.loc[overdf["match_id"] == match_id, col].values[0]
    
    if "shot.Goal" in new_df.columns:
        new_df.loc[:, "NPgoal"] = new_df.loc[:, "shot.Goal"].fillna(0) - new_df.loc[:, "shot.penalty.goal"].fillna(0)
    
    home_team = overdf.loc[overdf["match_id"] == match_id, "home_team.home_team_id"].values[0]
    away_team = overdf.loc[overdf["match_id"] == match_id, "away_team.away_team_id"].values[0]
    new_df.loc[:, "home"] = 0
    
    if overdf.loc[overdf["match_id"] == match_id, "competition.competition_id"].values[0] not in [72, 43]: 
        new_df.loc[new_df["team.id"] == home_team, "home"] = 1
        new_df.loc[new_df["team.id"] == away_team, "home"] = -1
    return new_df
```

```python
def playerSummary(match_ids, overdf):
    listSummary = []
    times = [time.time()]
    for i in range(0, len(match_ids)):
        match_id = match_ids[i]
        times.append(time.time())
        roundtime = round(times[-1] - times[-2], 2)
        print("Match number {} out of {}: {} seconds".format(i, len(match_ids), roundtime))
        playerdf = playerMatchSummary(match_id, overdf = overdf)
        listSummary.append(playerdf)
    Summaries = pd.concat(listSummary)

    unsums = ["player_id", "player_name", "player_nickname", "jersey_number", "country.id", "country.name", "team.name", 
              "team.id", "Name", "shortName", "player.id", "time.start", "time.end", 
              "opponent.id", "opponent.name", "match_id", "match_date", "competition.competition_id", 
              "competition.competition_name", "competition.country_name", "season.season_id", "season.season_name"]
    sums = [n for n in Summaries.columns if n not in unsums]
    df = Summaries.groupby(["Name", "team.name", "competition.competition_name", "season.season_name"])[sums].sum().reset_index()
    
    df.loc[:, "possession%"] = df.loc[:, "possession"] / df.loc[:, "possessions"]
    df.loc[:, "xG/shot"] =  df.loc[:, "shot.xG"] / df.loc[:, "shot.count"]
    df.loc[:, "NPxG/shot"] =  df.loc[:, "shot.NPxG"] / df.loc[:, "shot.NPcount"]
    
    for duel in "type.Interception", "duel.Tackle":
        d = duel.split(".")[-1]
        df.loc[:, "PAdj." + d] = df.loc[:, duel] * 2/(1 + np.e ** ( -10 * (df.loc[:, "possession%"] - 0.5)))
    return df
    
```

```python
def per90(df):
    spare = ["Name", "team.name", "competition.competition_name", "season.season_name", "time.played", 
             "time.start", "time.end"]
    cols = [n for n in df.columns if n not in spare]
    for col in cols:
        df[col] = df[col] * 90 / (df["time.played"] + 0.000001)
    return df
```

```python
def SumBar(df, stats, minutes = 0):
    df = df.copy()
    fig, ax = plt.subplots(figsize = (9, 12))
 
    for side in ["top", "bottom",  "right"]:
        ax.spines[side].set_visible(False) 

    for stat in stats:
        df.loc[:, stat + "90"] = df.loc[:, stat] * 90 / df.loc[:, "time.played"] + 0.00001
    df.loc[:, "stats"] = df.loc[:, [n + "90" for n in stats]].sum(axis = 1)
    gdf = df.loc[df["time.played"] > minutes, :]
    cdf = gdf.sort_values("stats", ascending = False)[:30]
    cdf.loc[:, "zeroes"] = 0

    x = [cdf.loc[:, "zeroes"]] + [cdf.loc[:, stat + "90"] for stat in stats]
    y = list(reversed(range(len(cdf))))

    tests, sub, lab = {}, [],  []
    cols = ["Name", "team.name", "competition.competition_name", "season.season_name"]
    for col in cols:
        tests[col] = cdf.loc[:, col].unique()
        if len(tests[col]) == 1:
            sub.append(tests[col][0])
        else:
            lab.append(col)
         
    if len(lab) == 0:
        cdf["labels"] = list(zip(*[cdf[n] for n in cols if n not in ["Name", "team.name"]]))
    else:
        cdf["labels"] = list(zip(*[cdf[n] for n in lab]))
        
    for i in range(len(stats)):
        plt.barh(y = y, width = x[i+1], height = 0.5, left = sum(x[:(i+1)]), 
                 tick_label = cdf["labels"], label = stats[i])

    x, new_y = cdf.stats.max()/2, max(y) + 1
    if minutes > 0:
        new_y += 0.5
        plt.text(x = x, y = new_y, s = "Minimum: {} minutes played".format(minutes), horizontalalignment = "center")
    if len(sub) > 0:
        new_y += 0.5
        plt.text(x = x, y = new_y, s = ", ".join(sub), horizontalalignment = "center")

    plt.text(x = x, y = new_y + 0.5 , s = " + ".join(stats), horizontalalignment = "center", fontsize = 12)
    
    plt.grid(True)
    ax.legend(loc = "lower right")
    plt.tight_layout()
    
    return fig, ax, plt
```

```python
def Scatter(df, stats, minutes = 0):
    fig, ax = plt.subplots(figsize = (10, 10))   

    for side in ["top", "bottom",  "right", "left"]:
        ax.spines[side].set_visible(False) 
    
    for stat in stats:
        if "/" in stat or "%" in stat:
            df.loc[:, stat + "90"] = df.loc[:, stat]
        else:
            df.loc[:, stat + "90"] = df.loc[:, stat] * 90 / df.loc[:, "time.played"] + 0.00001

    gdf = df.loc[df["time.played"] >= minutes].copy()
    
    tests, sub, lab = {}, [],  []
    cols = ["Name", "team.name", "competition.competition_name", "season.season_name"]
    for col in cols:
        tests[col] = gdf.loc[:, col].unique()
        if len(tests[col]) == 1:
            sub.append(tests[col][0])
        else:
            if col not in ["Name", "team.name"]:
                lab.append(col)

    if len(lab) == 0:
        gdf["labels"] = list(zip(*[gdf[n] for n in cols if n not in ["Name", "team.name"]]))
    else:
        gdf["labels"] = list(zip(*[gdf[n] for n in lab]))

    for label in gdf["labels"].unique():
        x = gdf.loc[gdf["labels"] == label, stats[0] + "90"]
        y = gdf.loc[gdf["labels"] == label, stats[1] + "90"]
        plt.scatter(x = x, y = y, s = 30, label = label, edgecolor = "k", alpha = 0.8, linewidth = 0.5)
    
    q = 0.95
    x = gdf.loc[:, stats[0] + "90"]
    y = gdf.loc[:, stats[1] + "90"]
    adf = gdf.loc[(gdf[stats[0] + "90"] > x.quantile(q))|(gdf[stats[1] + "90"] > y.quantile(q)), :]
    names = adf.loc[:, "Name"].apply(ShortName)
    for i, name in enumerate(names):
        xa = adf.loc[:, stats[0] + "90"].values[i] + x.max()/100
        ya = adf.loc[:, stats[1] + "90"].values[i] - y.max()/60
        plt.annotate(s = name, xy = (xa, ya))
    
    plt.axvline(c = "k")
    plt.axhline(c = "k")
        
    plt.xlabel(stats[0] + "90")
    plt.ylabel(stats[1] + "90")
    
    x, new_y, elev = x.max()/2, max(y) * 1.1, max(y)/25
    if minutes > 0:
        new_y += elev
        plt.text(x = x, y = new_y, s = "Minimum: {} minutes played".format(minutes), horizontalalignment = "center")
    if len(sub) > 0:
        new_y += elev
        plt.text(x = x, y = new_y, s = ", ".join(sub), horizontalalignment = "center")

    plt.text(x = x, y = new_y + elev , s = " against ".join(stats), horizontalalignment = "center", fontsize = 12)
    
        
    plt.grid(True)
    ax.legend(loc = "upper right")
    plt.tight_layout()
    
    return fig, ax, plt
```
